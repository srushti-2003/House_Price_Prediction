```python
data.describe()
```

```python

# Find the correlations between the features and the target variable (SalePrice)
corr_matrix = data.corr()
corr_saleprice = corr_matrix['SalePrice'].sort_values(ascending=False)

# Print the top 10 most correlated features
print(corr_saleprice[:7])
```

```python
x=data.loc[:,corr_matrix['SalePrice'].sort_values(ascending=False)[1:7].index]
x
```

```python
y=data['SalePrice']
y
```

```python
for col in data.columns[1:]:
    if data[col].dtype == 'object':
        if data[col].isnull().sum()!=0:
            data[col].fillna('missing',inplace=True)
    if data[col].dtype != 'object':
        if data[col].isnull().sum()!=0:
            data[col].fillna(0,inplace=True)
data
```

```python
from feature_engine.selection import DropConstantFeatures,DropCorrelatedFeatures,DropDuplicateFeatures
from sklearn.pipeline import Pipeline
pip=Pipeline([('Constant',DropConstantFeatures(tol=0.99)),('Duplicate',DropDuplicateFeatures()),('Correlated',DropCorrelatedFeatures())])
x_new=pip.fit_transform(x)
```

```python
x_new.shape
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for col in x_new.columns[1:]:
    if x_new[col].dtype == 'object':
        x_new[col]=l.fit_transform(x_new[col])
```

```python
x_new = x
x_new.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error
model=RandomForestRegressor()
x_train,x_test,y_train,y_test=train_test_split(x_new,y,train_size=0.7,random_state=0)
```

```python
model.fit(x_train,y_train)
pred=model.predict(x_test)
```

```python
r2_score(y_test,model.predict(x_test))
```

```python
mean_absolute_error(y_test,model.predict(x_test))
```
